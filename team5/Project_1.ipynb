```python
# Request the dataset via rest API
import requests
response = requests.get('https://covid19.th-stat.com/api/open/timeline')
```

```python
# Extract the json format data from the response
import pprint
jsondata = response.json()
pp = pprint.PrettyPrinter(indent=1)
pp.pprint(jsondata)
```

```python
# Extract only important part of jsondata for data analytics
raw_data = jsondata['Data']
pp.pprint(raw_data)
```

```python
raw_data[0]
```

```python
# Prepare the data for converting to the pandas dataframes
data = {}
features = list(raw_data[0].keys())

for feat in features:
    data[feat] = []
    
for i in range(len(raw_data)):
    for feat in raw_data[i].keys():     
        data[feat].append(raw_data[i][feat])
```

```python
data
```

```python
# Convert the json format data to pandas dataframe format
import pandas as pd
df = pd.DataFrame(data)
df
```

```python
features
```

```python
# Drop some beginning rows which the number of people infected with COVID19 is still too small (not pandemic or not too danger yet) 
outlier_index = df[df['Confirmed'] < 3000 ].index
```

```python
df.drop(outlier_index , inplace=True)
```

```python
# Drop most of the features since we state that we don't know anything in the future except the Date.
df.drop(features[1:5], axis=1, inplace=True)
df.drop(features[6:], axis=1, inplace=True)
df
```

```python
# Separate the 2020 and 2021 dataframe for being training data and test data respectively
split_index = df[df['Date']=='01/01/2021'].index.values
df.loc[split_index]
```

```python
df20 = df.iloc[:365-128]     # trainig data
df21 = df.iloc[365-128:]     # test data
```

```python
df20.index = [x for x in range(0, len(df20))]
df21.index = [x for x in range(0, len(df21))]
```

```python
# Convert date to a number of days for the purpose of linear regression.
for i in range(0, len(df20)):
    df20.loc[i, 'Date'] = i+1
```

```python
# Apply Machine Learning to predict the cumulative number of people infected with COVID19 in 2021 (Label is 'Confirmed')
# Choose Linear Regression Model and train the model with the data from 2020
y_train = df20['Confirmed']
x_train = df20.drop('Confirmed', axis=1)
```

```python
y_train
```

```python
x_train
```

```python
#fitting our model to training data and test data
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
print(x_train.columns)
print(model.intercept_)
print(model.coef_)
```

```python
# predict the cumulative number of people infected with COVID19 in 2021 with 2020 Linear Regression Model
y_test = df21['Confirmed']
x_test = df21.drop('Confirmed', axis=1)
y_pred = model.predict(x_test)
```

```python
import math
import sklearn
mse = sklearn.metrics.mean_squared_error(y_test, y_pred)
print(math.sqrt(mse))
```

```python
# R^2
print( model.score(x_test,y_test) )
```
